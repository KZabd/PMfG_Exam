```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
```

```python
temp1 = rasterio.open("/Users/user/Studia_RSG/S2/Python_Matlab_for_Geoscience/Exam/t5_lst2023_Jun_Aug.tif").read(1)
temp2 = rasterio.open("/Users/user/Studia_RSG/S2/Python_Matlab_for_Geoscience/Exam/t5_lst2024May.tif").read(1)
ndvi1 = rasterio.open("/Users/user/Studia_RSG/S2/Python_Matlab_for_Geoscience/Exam/t5_ndvi2024_Jul_Aug.tif").read(1)
ndvi2 = rasterio.open("/Users/user/Studia_RSG/S2/Python_Matlab_for_Geoscience/Exam/t5_ndvi2024May.tif").read(1)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 8))
cmtemp = "RdBu_r"
cmndvi = "RdYlGn_r"
cmdiff = "Reds"

show(temp1, ax = axes[0, 0], cmap = cmtemp)
axes[0, 0].set_title("Land Surface Temperature Jun-Aug 2023")
axes[0, 0].axis("off")

show(ndvi1, ax = axes[0, 1], cmap = cmndvi)
axes[0, 1].set_title("NDVI Jul-Aug 2024")
axes[0, 1].axis("off")

show(temp2, ax = axes[1, 0], cmap = cmtemp)
axes[1, 0].set_title("Land Surface Temperature May 2024")
axes[1, 0].axis("off")

show(ndvi2, ax = axes[1, 1], cmap = cmndvi)
axes[1, 1].set_title("NDVI May 2024")
axes[1, 1].axis("off")

tempdiff = temp2 - temp1
show(tempdiff, ax = axes[2, 0], cmap = cmdiff)
axes[2, 0].set_title("Land Surface Temperature difference")
axes[2, 0].axis("off")

ndvidiff = ndvi2 - ndvi1
show(ndvidiff, ax = axes[2, 1], cmap = cmdiff)
axes[2, 1].set_title("NDVI difference")
axes[2, 1].axis("off")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes[0, 0].hist(temp1.flatten(), bins=100, color='blue', alpha=0.7)
axes[0, 0].set_title("Histogram of Land Surface Temperature Jun-Aug 2023")
axes[0, 0].set_xlabel("Temperature")

axes[0, 1].hist(ndvi1.flatten(), bins=100, color='blue', alpha=0.7)
axes[0, 1].set_title("Histogram of NDVI Jul-Aug 2024")
axes[0, 1].set_xlabel("NDVI")

axes[1, 0].hist(temp2.flatten(), bins=100, color='blue', alpha=0.7)
axes[1, 0].set_title("Histogram of Land Surface Temperature May 2024")
axes[1, 0].set_xlabel("Temperature")

axes[1, 1].hist(ndvi2.flatten(), bins=100, color='blue', alpha=0.7)
axes[1, 1].set_title("Histogram of NDVI May 2024")
axes[1, 1].set_xlabel("NDVI")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].scatter(ndvi1.flatten(), temp1.flatten(), alpha=0.3, color='blue')
axes[0].set_xlabel("NDVI")
axes[0].set_ylabel("Temperature")
axes[0].set_title("NDVI to Temperature Jun-Aug 2023")

axes[1].scatter(ndvi2.flatten(), temp2.flatten(), alpha=0.3, color='blue')
axes[1].set_xlabel("NDVI")
axes[1].set_ylabel("Temperature")
axes[1].set_title("NDVI to Temperature May 2024")

plt.show()
```
